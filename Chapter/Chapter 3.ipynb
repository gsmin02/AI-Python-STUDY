```python
#리스트3.1
import numpy as np
import matplotlib.pyplot as plt

print("변수는 x와 y 등의 알파벳을 이용해서 자주 나타냅니다. 상수는 1, 2.3, -5 등의 수치로 표기합니다. 또한, a 나 b 등의 알파벳, α 나 β 등의 그리스 문자로도 자주 나타냅니다. 아래는 y = ax의 식에 따라 직선이 그려졌습니다. 이 경우 x와 y 값은 변화해도 a 값은 변화하지 않습니다.")

a = 1.5 # a : 상수
x = np.linspace(-1, 1) # x: 변수 -1부터 1의 범위
y = a * x # y: 변수

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
#리스트3.7
import numpy as np
import matplotlib.pyplot as plt

print("수식 y=x^a를 코드로 구현합니다. Python에서의 거듭제곱은 **로 기술합니다.")

def my_func(x):
    a = 3
    return x**a # x의 a제곱

x = np.linspace(0, 2)
y = my_func(x) # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
#리스트3.14
import numpy as np
import matplotlib.pyplot as plt

print("y = x^3 - 2x^2 - 3x + 4 수식을 그래프로 그려봅시다.")

def my_func(x):
    return x**3 - 2*x**2 - 3*x + 4

x = np.linspace(-2, 2)
y = my_func(x) # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
#리스트3.15
import numpy as np
import matplotlib.pyplot as plt

print("NumPy의 sin() 함수와 cos() 함수를 사용하는데 인수의 단위는 라디안입니다. 원주율은 np.pi로 얻을 수 있습니다.")

def my_sin(x):
    return np.sin(x) # sin(x)

def my_cos(x):
    return np.cos(x) # cos(x)

x = np.linspace(-np.pi, np.pi) # -pi부터 pi(라디안)까지
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend() # 범례

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

## LaTeX의 기초
Jupyter Notebook은 LaTeX이라는 문서 처리 시스템을 사용해서 수식을 표현할 수 있습니다.  
Jupyter Notebook의 셀타입을 마크다운으로 하고 입력해 실행합니다.

$$y=2x+1$$

마크다운의 셀에는 두 겹의 ＄ 안에 수식을 나타내는 LaTeX 코드를 기술합니다.  
또한, 문장 중에 수식을 삽일할 때는 ＄y=2x+1＄처럼 한 겹의 ＄ 안에 코드를 기술합니다.
  
### 첨자와 거듭제곱
^ 와 _ 의 기호를 사용해서 오른쪽 아래의 첨자와 오른쪽 위의 거듭제곱을 표현할 수 있습니다. 첨자가 여러 개 있는 경우는 { } 로 갑쌉니다  
  
$a_1$ -> a_1  
$a_{ij}$ -> a_{ij}  
$b^2$ -> b^2  
$b^{ij}$ -> b^{ij}  
$c_1^2$ -> c_1^2  
  
### 다항식
첨자나 거듭제곱을 이용해서 다항식을 만들 수 있습니다.  
- 수식의 예: $y = x^3+2x^2+x+3$
- LaTeX의 기술: y = x^3+2x^2+x+3
  
### 제곱근
\sqrt를 이용해 √ 기호를 기술할 수 있습니다.
- 수식의 예: $y=\sqrt x$
- LaTeX의 기술: y=\sqrt x
  
### 삼각함수
\sin이나 \cos을 이용해 삼각함수를 입력할 수 있습니다.
- 수식의 예: $y=\sin x$
- LaTeX의 기술: y=\sin x
  
### 분수
\frac{}{}을 이용해 분수를 기술 할 수 있습니다.
- 수식의 예: $y=\frac{17}{24}$
- LaTeX의 기술: y=\frac{17}{24}
  
### 총합
\sum을 이용해 $\sum$ 기호를 기술 할 수 있습니다.
- 수식의 예: $y=\sum_{k=1}^n a_k$
- LaTeX의 기술: y=\sum_{k=1}^n a_k
  
### 총곱
\prod을 이용해 $\prod$ 기호를 기술할 수 있습니다.
- 수식의 예: $y=\prod_{k=1}^n a_k$
- LaTeX의 기술: y=\prod_{k=1}^n a_k
  
## 활용
다음의 수식을 Jupyter Notebook의 셀에 LaTeX 형식으로 기술할 수 있습니다.
$$y=x^3 + \sqrt x + \frac{a_{ij}}{b_{ij}^4} - \sum_{k=1}^n a_k$$
- LaTeX의 기술: y=x^3 + \sqrt x + \frac{a_{ij}}{b_{ij}^4} - \sum_{k=1}^n a_k

```python
#리스트3.31
import numpy as np

print("절댓값은 NumPy의 abs() 함수를 사용해 구할 수 있습니다.")

x = [-5, 5, -1.28, np.sqrt(5), np.pi/2]
# 여러 값을 리스트에 저장한다.
print(np.abs(x)) # 절댓값을 구한다.
```

```python
#리스트3.32
import numpy as np
import matplotlib.pyplot as plt

print("절댓값의 이미지를 파악하기 위해서 함수의 절댓값을 구해서 그래프로 표시해 봅시다.")

x = np.linspace(-np.pi, np.pi) # -pi부터 pi(라디안)까지
y_sin = np.abs(np.sin(x)) # sin() 함수의 절댓값을 취한다
y_cos = np.abs(np.cos(x)) # cos() 함수의 절댓값을 취한다

plt.scatter(x, y_sin, label="sin")
plt.scatter(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
#리스트3.34
import numpy as np
import matplotlib.pyplot as plt

print("2차함수의 절댓값을 취하고, 그래프가 어떻게 변화하는지 확인해 봅시다.")

x = np.linspace(-4, 4)
y = np.abs(x**2 - 4) # 이 2차함수의 절댓값을 취한다.

plt.scatter(x, y)

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```
