```python
#리스트2.15
print("튜플은 리스트와 같이 여러 개의 값을 한데 모아서 다룰 때에 유용하지만 요소의 추가, 삭제, 교체 등은 할 수 없습니다. 튜플은 값(요소) 전체를 ()로 감싸고, 각 요소는 ,로 구분합니다. 요소를 변경하지 않을 때에는 리스트보다 튜플을 사용하는 편이 좋습니다.")

a = (1, 2, 3, 4, 5) # 튜플의 작성
b = a[2] # 인덱스가 2인 요소를 획득
print(b)
```

```python
#리스트2.51
import numpy as np
import matplotlib.pyplot as plt

print("pyplot을 사용하여 직선을 그립니다. NumPy의 linspace() 함수로 x 좌표의 데이터를 배열로 생성하고 이것에 값을 곱해서 y 좌표로 합니다. 그리고 pyplot의 plot() 함수로 x 좌표, y 좌표의 데이터를 플롯하고, show() 함수로 그래프를 표시합니다.")

x = np.linspace(-5, 5) # -5부터 5까지
y = 2 * x # x에 2를 곱해서 y좌표로 한다

plt.plot(x, y)
plt.show()
```

```python
#리스트2.52
import numpy as np
import matplotlib.pyplot as plt

print("축의 라벨, 그래프의 타이틀, 그리드의 표시, 범례와 선의 스타일을 표시해 그래프의 외관을 풍부하게 합니다.")

x = np.linspace(-5, 5)
y_1 = 2 * x
y_2 = 3 * x

# 축의 라벨
plt.xlabel("x value", size = 14)

plt.ylabel("y value", size = 14)

# 그래프의 다이틀
plt.title("My Graph")

# 그리드 표시
plt.grid()

# 플롯 시에 범례와 선의 스타일을 지정
plt.plot(x, y_1, label ="y1")
plt.plot(x, y_2, label ="y2", linestyle="dashed")
plt.legend() # 범례를 표시

plt.show()
```

```python
#리스트 2.53
import numpy as np
import matplotlib.pyplot as plt

print("scatter() 함수로 산포도를 표시할 수 있습니다. x 좌표, y 좌표로부터 산포도를 그립니다.")


x = np.array([1.2, 2.4, 0.0, 1.4, 1.5, 0.3, 0.7]) # x 좌표
y = np.array([2.4, 1.4, 1.0, 0.1, 1.7, 2.0, 0.6]) # y 좌표

plt.scatter(x, y) # 산포도의 슬롯
plt.grid()
plt.show()
```

```python
#리스트2.54
import numpy as np
import matplotlib.pyplot as plt

print("hist() 함수로 히스토그램을 표시할 수 있습니다. 히스토그램에서는 각 범위의 값의 빈도가 카운트된 직사각형의 기둥으로 나타납니다. 배열 data에 있어서의 각 값의 빈도를 카운트한 히스토그램으로 표시합니다.")

data = np.array([0, 1, 1, 2, 2, 2, 3, 3, 4, 5, 6,6, 7, 7, 7, 8, 8, 9])

plt.hist(data, bins=10) # 히스토그램 bins는 기둥의 수
plt.show()
```

```python
#리스트2.56
import numpy as np
import matplotlib.pyplot as plt

print()

x = np.linspace(-3, 3) # x의 범위를 지정
y_1 = 1.5 * x # x에 연산을 시행하는 y_1으로 한다.
y_2 = -2 * x + 1 # x에 연산을 시행하는 y_2으로 한다.

# 축의 라벨
plt.xlabel("x value", size=14)
plt.ylabel("y value", size=14)

# 그래프의 타이틀
plt.title("My Graph")

# 그리드 표시
plt.grid()

# 플롯 범례와 선의 스타일을 지정
plt.plot(x, y_1, label="y1")
plt.plot(x, y_2, label="y2", linestyle="dashed")
plt.legend() # 범례를 표시

plt.show()
```

```python
#리스트3.1
import numpy as np
import matplotlib.pyplot as plt

print("변수는 x와 y 등의 알파벳을 이용해서 자주 나타냅니다. 상수는 1, 2.3, -5 등의 수치로 표기합니다. 또한, a 나 b 등의 알파벳, α 나 β 등의 그리스 문자로도 자주 나타냅니다. 아래는 y = ax의 식에 따라 직선이 그려졌습니다. 이 경우 x와 y 값은 변화해도 a 값은 변화하지 않습니다.")

a = 1.5 # a : 상수
x = np.linspace(-1, 1) # x: 변수 -1부터 1의 범위
y = a * x # y: 변수

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```
